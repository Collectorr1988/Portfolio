–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–∏–±—ã–ª–∏ —Å –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏ –ø—Ä–∏–º–µ–Ω–∏–º –µ–µ –∫ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ

```python
def proceeds_calculate (row): 

#—Å–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –∏–∑ df_merge3 –¥–ª—è –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π

    row_tariff=row['tariff']
    row_sum_duration_calls=row['sum_duration_calls']
    row_message_sum=row['message_sum']
    row_mb_used=row['mb_used']
    
#—Å–æ–∑–¥–∞–¥–∏–º —Å–ª–æ–≤–∞—Ä—å —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞

    smart={'messages_included_s':50, 'mb_per_month_included_s':15360, 'minutes_included_s':500, 'rub_monthly_fee_s':550 
       ,'rub_per_gb_s':200, 'rub_per_message_s':3, 'rub_per_minute_s':3}
    
    ultra={'messages_included_u':1000, 'mb_per_month_included_u':30720, 'minutes_included_u':3000, 'rub_monthly_fee_u':1950
       ,'rub_per_gb_u':150, 'rub_per_message_u':1, 'rub_per_minute_u':1}
    
#–µ—Å–ª–∏ –∑–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–±—Ü–µ tariff –≤ —Ç–∞–±–ª–∏—Ü–µ=smart —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –ø—Ä–∏–±—ã–ª—å –¥–ª—è smart, –µ—Å–ª–∏ ultra, —Ç–æ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –ø—Ä–∏–±—ã–ª—å –¥–ª—è ultra
#–≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏

    if  row_tariff == 'smart':
        minutes_s=(row_sum_duration_calls-smart['minutes_included_s'])*smart['rub_per_minute_s']
        messages_s=(row_message_sum-smart['messages_included_s'])*smart['rub_per_message_s']
        internet_s=(row_mb_used-smart['mb_per_month_included_s'])*0.000977*smart['rub_per_gb_s']
        
#—É—á—Ç–µ–º –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ—Ö–æ–¥–∞ –ø–æ –∫–∞–∂–¥–æ–π —É—Å–ª—É–≥–µ –∏ –≤ —ç—Ç–æ–º —Å–ª—É—á–∞–µ –±—É–¥–µ—Ç –≤–æ–∑—Ä–∞—â–∞—Ç—å –∞–±–æ–Ω–µ–Ω—Ç—Å–∫—É—é –ø–ª–∞—Ç—É

        if minutes_s>=0:
             proceeds_minutes_s=minutes_s
        else:
             proceeds_minutes_s=0
            
        if messages_s>=0:
             proceeds_messages_s=messages_s
        else:
             proceeds_messages_s=0
            
        if internet_s>=0:
             proceeds_internet_s=internet_s
        else:
             proceeds_internet_s=0
        
        proceeds=proceeds_minutes_s+proceeds_messages_s+proceeds_internet_s+smart['rub_monthly_fee_s']
        
        return(proceeds)
    
# –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ ultra
    
    else:
        minutes_u=(row_sum_duration_calls-ultra['minutes_included_u'])*ultra['rub_per_minute_u']
        messages_u=(row_message_sum-ultra['messages_included_u'])*ultra['rub_per_message_u']
        internet_u=(row_mb_used-ultra['mb_per_month_included_u'])*0.000977*ultra['rub_per_gb_u']
               
        
        if minutes_u>=0:
            proceeds_minutes_u=minutes_u
        else:
            proceeds_minutes_u=0
            
        if messages_u>=0:
            proceeds_messages_u=messages_u
        else:
            proceeds_messages_u=0
            
        if internet_u>=0:
            proceeds_internet_u=internet_u
        else:
            proceeds_internet_u=0
            
        proceeds=proceeds_minutes_u+proceeds_messages_u+proceeds_internet_u+ultra['rub_monthly_fee_u']    
        return(proceeds)
        
#–ø—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫–∏—é –∫ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ

df_merge3['proceeds']=df_merge3.apply(proceeds_calculate, axis=1)
display(df_merge3)
```
